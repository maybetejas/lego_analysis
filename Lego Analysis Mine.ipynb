```python
import pandas as pd
```

```python
pt = pd.read_csv('parent_themes.csv')
pt.head()
```

```python
lsets = pd.read_csv('lego_sets.csv')
lsets
```

# What % of licensed theme sets were star wars

```python
mergedDf = pd.merge(pt, lsets, left_on='name', right_on='parent_theme', how='inner')
del mergedDf['name_x']
mergedDf
```

```python
licensedDf = mergedDf[mergedDf['is_licensed']]
n = len(licensedDf)
```

```python
starWarsDf = mergedDf[mergedDf['parent_theme'].str.contains('Star Wars', case=False)]
x = len(starWarsDf)
```

```python
x / n
```

#### Answer: 45.72%

```python
licensedSorted = licensedDf.sort_values('year')
licensedSorted['count'] = 1
g = licensedSorted.groupby(['year', 'parent_theme']).sum().reset_index()
g
```

```python
mx = g.sort_values('count', ascending=False).drop_duplicates(['year'])
res = mx.sort_values('year')
res
```

### Answer: 2017 => Superheroes

## How many unique sets were released every year

```python
licensedSorted
```

```python
g2 = licensedSorted.groupby(['year', 'set_num']).size().reset_index()
g2.head()
```

### Answer

```python
g4 = g2.groupby('year')['set_num'].nunique().reset_index(name='total_count')
g4.head()
```
